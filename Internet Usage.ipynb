## 1). Which Country makes the most money from Internet Consumption

```python
data.head()  ## overview our data
```

```python
## We may need to drop columns we dont need to answer this question
data_q1 =data.drop(columns =['Country code','Continental region','Subregion'])
data_q1
```

```python
### We are going to create a new column which is going to give us the cost made by countries
### It is going to be the product of internet users by the average price of 1GB

data_q1['cost_of_internet'] = data_q1['Average price of 1GB (USD)']* data_q1['Internet users']
data_q11=data_q1.sort_values('cost_of_internet', ascending=False).head(5)
data_q11
```

```python
x=data_q11['country_name']
y=data_q11['cost_of_internet']
plt.bar(x,y, label='cost of internet')
plt.title('Country Revenue from Internet Users')
plt.legend();
```

```python
## We can deduce that the United States makes more than anyother country from internet cost
```

## 2. Is there any Relationship between the cost of Internet and Country Population?

```python
## We will need a component bar chart to conclude on this question
```

```python
k_range=np.arange(1,6)
z=data_q11['Population']
plt.bar(k_range,y,width=0.4,label="Internet Cost");
plt.bar(k_range+0.4,z,width=0.4,label="population");
plt.xticks(k_range+0.2,x);
plt.legend();
```

```python
### Seemingly the top 5 countries make equivalently twice the size of their population as revenue apart from India and China
### We can therefore say cost of Internet in India and China is relatively cheap compared to their population
### We can get another question from this, why the cost of internet is cheap in Indian and China
```

## 4. Is there any form of correlation/Relationship in the variables

```python
data.head()
```

```python
data.columns
```

```python
### Firstly, we are going to plot a create a correlation table by dropping tables we dont need
corr_table=data.drop(columns=['Country code','country_name','Continental region','Subregion','Region','Average price of 1GB (USD  at the start of 2021)','Average price of 1GB (USD at the start of 2020)'])
corr_table
```

```python
## Plotting the correlation Table
plt.figure(figsize=(14,6))
sns.heatmap(corr_table.corr(),annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno',linewidths=1,linecolor='Black').set(title='Correlation Map');
```

```python
## The only strong correlation we can observe is the r/ship between the cheapest 1Gb and the average price of 1Gb
## This is so because the average price of 1Gb is the mean of all the 1Gb prices - the cheapest and most expensive inclusively
## We can plot a scatter plot below to see the r/ship between this 2 variables

plt.scatter(data['Cheapest 1GB for 30 days (USD)'],data['Average price of 1GB (USD  at the start of 2021)']);
```
